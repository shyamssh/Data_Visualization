# Univariate Data Visualisation

**Univariate visualisation means that the visualisation is done using only one variable**
**if a dataset is considered then the visualisation around a single feature of the dataset is univariate visualisation.**

# The Histograms and Bar charts 

**A histogram represents the frequency distribution of continuous variables. Conversely, a bar graph is a diagrammatic comparison of discrete variables.**<br>
**Histogram presents numerical data whereas bar graph shows categorical data.**

# Bar Charts 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("Hospitals_and_Beds_statewise.csv")
df.columns = ['states', 'PHC', 'CHC', 'SDH', 'DH', 'Total', 'beds']
```

```python
df = df[['states','Total', 'beds']]
df.Total = pd.to_numeric(df.Total, errors='coerce').fillna(0).astype(np.int64)
df.beds = pd.to_numeric(df.beds, errors='coerce').fillna(0).astype(np.int64)
```

```python
df.sort_values(by=['Total'], inplace=True, ascending=False)
```

```python
states = df['states'].tolist()
hospitals = df['Total'].tolist()
states = states[0:8]
hospitals = hospitals[0:8]
x_pos = np.arange(len(states))

plt.bar(x_pos, hospitals, align='center', alpha=0.8, color='c')

plt.xticks(x_pos, states, rotation=70)
plt.ylabel('Hospital Count')
plt.title('Statewise hospitals in India')
```

# Horizontal Bar Charts 

```python
states = df['states'].tolist()
hospitals = df['Total'].tolist()
states = states[0:8]
hospitals = hospitals[0:8]
y_pos = np.arange(len(states))
options = {'align':'center', 'alpha':0.8, 'color':'c'}

plt.barh(y_pos, hospitals, **options)

plt.yticks(y_pos, states)
plt.xlabel('Hospital Count')
plt.title('Statewise hospitals in India')
```

# Bar Charts Comparison 

```python
num_states = 8

states = df['states'].tolist()
hospitals = df['Total'].tolist()
beds = df['beds'].tolist()

states = states[0:num_states]
hospitals = hospitals[0:num_states]
beds = beds[0:num_states]

# create plot
fig, ax = plt.subplots()
index = np.arange(num_states)
bar_width = 0.40
opacity = 0.8

rects1 = plt.bar(index, hospitals, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Hospital Count')

rects2 = plt.bar(index + bar_width, beds, bar_width,
                 alpha=opacity,
                 color='g',
                 label='beds Count')

plt.xlabel('hospitals and beds count')
plt.ylabel('Number of units')
plt.title('Statewise hospitals and beds in India')
plt.xticks(index + bar_width/2, states, rotation=70)
plt.legend()
plt.tight_layout()
```

# Histograms
**In a histogram, each bar groups numbers into ranges**

```python
df_bank = pd.read_csv("Churn_Modelling.csv")
# Filter the data where the customer have exicited the bank.
df_bank = df_bank[df_bank['Exited'] == 1]
df_bank.head()
```

```python
age = df_bank['Age'].tolist()
fig,ax = plt.subplots()
ax.hist(age, bins=10)
ax.set_xlabel('Age Range')
ax.set_ylabel('Age')
ax.set_title(r'Distribution of Age of the customers Exited')
```
